```python
#import libraries
```

```python
import  json 
import urllib
import  urllib.parse
import urllib.request
```

```python
serviceurl='http://maps.googleapis.com/maps/api/geocode/json?'
```

```python
url=serviceurl+urllib.parse.urlencode({'sensor':'false','address':area_zone})
print(url)
```

```python
uh=urllib.request.urlopen(url)
```

```python
uh
```

```python
data=uh.read()
data
""" it  seems  it is  byte format"""
```

```python
js=json.loads(str(data.decode('utf8')))
js
```

```python
print(json.dumps(js,indent=4))
```
